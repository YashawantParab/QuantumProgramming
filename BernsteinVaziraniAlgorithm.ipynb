# Bernstein-Vazirani Algorithm

```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secretnumber = '101001'
```

```python
circuit = QuantumCircuit(6+1, 6) #6 because of the 6 digits 101001 +1 qubit and 6 classical bits on which we'll store result. 
```

```python
# apply hadamard gate to first  6 bits
```

```python
circuit.h([0,1,2,3,4,5]) 
```

```python
# for last bit add X gate  before hadamard gate
circuit.x(6)
circuit.h(6)
```

```python
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
# bulid the box that content secret number
# for every one in the secret number, put cX gate
```

### cX gate

```python
circuit.cx(5,6) #101001 5th place has 1 
circuit.cx(3,6) #101001 3rd place has 1 
circuit.cx(0,6) #101001 oth place has 1
```

```python
circuit.barrier()
```

```python
circuit.draw(output='mpl')
```

```python
circuit.h([0,1,2,3,4,5]) 
```

```python
circuit.draw(output='mpl')
```
